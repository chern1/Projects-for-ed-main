```python
import glob
import os
import codecs

path = r"C:\Users\MAx\OneDrive\Рабочий стол"
files = glob.glob(path + "/*.csv")


li = []

for filename in files:
    data = pd.read_csv(filename, encoding='utf-8', sep = ';', on_bad_lines='skip')
    li.append(data)

df = pd.concat(li, axis=0, ignore_index=True)
```

# Настойка таблицы

```python
df.dropna(how = 'any', inplace = True)
```

```python
df = df.drop(df.columns[[0,11,12,13,14]], axis=1)
```

```python
df['category_id'] = pd.to_numeric(df['category_id'], errors='coerce')
df['category_id'] = df['category_id'].astype(int)
df['category_id'] = df['category_id'].astype('category')
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m', errors='coerce')
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'], infer_datetime_format=True, errors='coerce', utc=True)
df['publish_time'] = df['publish_time'].dt.strftime("%d/%m/%y")
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

```python
df
```

```python
df['views'] = df['views'].astype(int)
df['likes'] = df['likes'].astype(int)
df['dislikes'] = df['dislikes'].astype(int)
df['comment_count'] = df['comment_count'].astype(int)
```
