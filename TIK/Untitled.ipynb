```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('top_1000_tiktokers.csv', index_col='Rank')
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Subscribers Count'] = df['Subscribers Count'].replace({"M": "*1E6", "K": "*1E3"}, regex=True).map(pd.eval).astype(int)
df['Views. Avg'] = df['Views. Avg'].replace({"M": "*1E6", "K": "*1E3"}, regex=True).map(pd.eval).astype(int)
df['Likes. Avg'] = df['Likes. Avg'].replace({"M": "*1E6", "K": "*1E3"}, regex=True).map(pd.eval).astype(int)
df['Comments. Avg'] = df['Comments. Avg'].replace({"M": "*1E6", "K": "*1E3"}, regex=True).map(pd.eval).astype(int)
df['Shares. Avg'] = df['Shares. Avg'].replace({"M": "*1E6", "K": "*1E3"}, regex=True).map(pd.eval).astype(int)
```
