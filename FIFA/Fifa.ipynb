```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv', index_col = 'Name')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[['Age', 'Nationality', 'Overall', 'Potential', 'Club', 'Value', 'Position', 'Wage', 'Height', 'Weight']]
df
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dropna(how = 'any', inplace = True)
df.sort_values('Weight')
```

```python
df['Value'] = df['Value'].replace({"€": "", "M": "*1E6", "K": "*1E3"}, regex=True).map(pd.eval).astype(int)
df['Wage'] = df['Wage'].replace({"€": "", "M": "*1E6", "K": "*1E3"}, regex=True).map(pd.eval).astype(int)
df['Weight'] = df['Weight'].replace({"lbs":""}, regex=True).astype(int)
df['Weight'] = (df['Weight']*0.43).round(1)
df['Position'] = df['Position'].astype('category')
df
```

```python
plt.subplot(1,2,1)
df['Overall'].value_counts().sort_index(ascending=True).plot(kind = 'bar', figsize=(15, 10))


plt.subplot(1,2,2)
wewe = df['Weight'].value_counts().sort_index(ascending=True).plot(kind = 'bar', figsize=(15, 10))
plt.show()
```
