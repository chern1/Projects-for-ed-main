```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Stores.csv')
```

```python
df.info()
```

```python
df
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.drop(df.columns[[0]], axis = 1, inplace = True)
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
condition_list=[(df['Store_Area']<=1250),(df['Store_Area']>1250) & (df['Store_Area']<=1750),(df['Store_Area']>1750)]
value_list=['Small','Medium','Large']
df['Type']=np.select(condition_list,value_list)
```

```python
sns.heatmap(df.corr(), cmap= sns.color_palette(palette="RdBu"),linewidths=0.5,annot=True)
```

```python
sns.set()
sns.pairplot(df,height = 5 ,kind ='reg',diag_kind='kde')
plt.show()
```
