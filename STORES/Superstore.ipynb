```python
df.columns = df.columns.str.replace(' ','_')   # remove spaces in columns name
```

```python
df.head(5)
```

```python
condition_list=[(df['Store_Area']<=1250),(df['Store_Area']>1250) & (df['Store_Area']<=1750),(df['Store_Area']>1750)]
value_list=['Small','Medium','Large']
```

```python
df['Type']=np.select(condition_list,value_list)
```

```python
df
```

```python
sns.scatterplot(df.Store_Area,df.Daily_Customer_Count,hue=df.Type)
plt.title('Area vs Customer Count');
```

Maximum number of Stores are having area in the range of 1400-1600 yard square and the daily customer count is also in the range of 700-900 people. As the area increases or decreases from the average area of 1485.409598 yard square, the customer count also reduces. Infact the lowest as well as highest daily customer count is seen in the 1400-1600 yard square area. So having an area too large as well as too small cannot reduce the customers.

```python
plt.hist(df.Store_Area,bins=np.arange(700,2300,100))
plt.title('Store Distribution by Area');
```

```python
sns.scatterplot(x=df.Items_Available,y=df.Daily_Customer_Count,hue=df.Type);
```

```python
df[['Store_Area','Items_Available']].sort_values('Items_Available')
```

```python
sns.scatterplot(x=df.Store_Area,y=df.Items_Available);
```

You can see that as the area of the store increases the Items available within the store also increases but that doesn't attract many customers. The stores with medium area have 1500-2000 Items avaiable in them and they are the ones that attract maximum customers only. So providing many items to customers in order to attract them can play a negative role as well. So the best way to gain more customers is by having a store with adequate area of around 1500 yard square and items in range of  1750.

```python
fig, axes =plt.subplots(2,3,figsize=(19,9))

axes[0,0].set_title('Daily Customer vs Store Sales')
sns.scatterplot(x=df.Daily_Customer_Count,y=df.Store_Sales,ax=axes[0,0],hue=df.Type)

axes[0,1].set_title('Items Available vs Store Sales')
sns.scatterplot(x=df.Items_Available,y=df.Store_Sales,ax=axes[0,1],hue=df.Type)

axes[0,2].set_title('Store Area vs Store Sales')
sns.scatterplot(x=df.Store_Area,y=df.Store_Sales,ax=axes[0,2],hue=df.Type)

axes[1,1].set_title('Majority of sales')
axes[1,1].hist(df.Store_Sales)

plt.tight_layout(pad=2)
```

When it comes to sales, we can see though the top sales of $ 116320 and $105150	is done by two large stores respectively, the majority of stores have a sales of around $40000-$70000. The shocking fact is that the top two stores also have a customer count of 860 and 980 respectively which is in average range of 600-1100 customers. This tells us the fact that on a average of 700-1000 customers are enough to increase the sales. Thus having a bigger store area with a lot of items can confuse the customers as well thus playing a negative role in increasing the sales. Also having a smaller area and less items can play an evenly negative role in increasing the sales.
